```python
import pygame
import time
import random

WIDTH = 600
HEIGHT = 600
BLOCK_SIZE = 20
FPS = 10

WHITE = (255, 255, 255)
GREEN = (0, 255, 0)
RED = (255, 0, 0)

BLOCK_SIZE = 10

class Snake:
    def __init__(self):
        self.size = 1
        self.elements = [[100, 100]]
        self.direction = 'right'
        self.is_add = False

    def draw(self, window):
        for element in self.elements:
            pygame.draw.rect(window, GREEN, [element[0], element[1], BLOCK_SIZE, BLOCK_SIZE])

    def add_to_snake(self):
        self.size += 1
        self.elements.append([0, 0])
        self.is_add = False

    def move(self):
        if self.direction == 'right':
            temp = self.elements[0][:]
            self.elements[0][0] += BLOCK_SIZE
            for i in range(1, self.size):
                _ = self.elements[i][:]
                self.elements[i] = temp
                temp = _

        if self.direction == 'left':
            temp = self.elements[0][:]
            self.elements[0][0] -= BLOCK_SIZE
            for i in range(1, self.size):
                _ = self.elements[i][:]
                self.elements[i] = temp
                temp = _

        if self.direction == 'up':
            temp = self.elements[0][:]
            self.elements[0][1] -= BLOCK_SIZE
            for i in range(1, self.size):
                _ = self.elements[i][:]
                self.elements[i] = temp
                temp = _

        if self.direction == 'down':
            temp = self.elements[0][:]
            self.elements[0][1] += BLOCK_SIZE
            for i in range(1, self.size):
                _ = self.elements[i][:]
                self.elements[i] = temp
                temp = _

        if self.is_add:
            self.add_to_snake()

class Food:
    def __init__(self):
        self.x = random.randint(0, (WIDTH - BLOCK_SIZE) // BLOCK_SIZE) * BLOCK_SIZE
        self.y = random.randint(0, (HEIGHT - BLOCK_SIZE) // BLOCK_SIZE) * BLOCK_SIZE

    def draw(self, window):
        pygame.draw.rect(window, RED, [self.x, self.y, BLOCK_SIZE, BLOCK_SIZE])


def check_collision(snake_elements, food):
    return snake_elements[0] == [food.x, food.y]

def game_over(snake):
    return (snake.elements[0][0] > WIDTH-BLOCK_SIZE or snake.elements[0][0] < 0 or 
            snake.elements[0][1] > HEIGHT-BLOCK_SIZE or snake.elements[0][1] < 0 or
            snake.elements[0] in snake.elements[1:])

def main():
    pygame.init()
    window = pygame.display.set_mode((WIDTH, HEIGHT))
    clock = pygame.time.Clock()

    snake = Snake()
    food = Food()

    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()

        keys = pygame.key.get_pressed()
        if keys[pygame.K_UP] and snake.direction != 'down':
            snake.direction = 'up'
        elif keys[pygame.K_DOWN] and snake.direction != 'up':
            snake.direction = 'down'
        elif keys[pygame.K_LEFT] and snake.direction != 'right':
            snake.direction = 'left'
        elif keys[pygame.K_RIGHT] and snake.direction != 'left':
            snake.direction = 'right'

        window.fill(WHITE)
        snake.move()
        snake.draw(window)
        food.draw(window)
        
        if check_collision(snake.elements, food):
            snake.is_add = True
            food = Food()

        if game_over(snake):
            pygame.quit()
            quit()

        pygame.display.flip()
        clock.tick(FPS)

if __name__ == "__main__":
    main()
```
